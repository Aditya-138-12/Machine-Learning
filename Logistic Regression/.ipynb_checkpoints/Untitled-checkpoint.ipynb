```python
# Here 1 -> Çerçevelik
# and  2 -> Ürgüp Sivrisi

df["Class"], class_labels = pd.factorize(df["Class"])

for idx, label in enumerate(class_labels, start = 0):
    print(f"{idx} -> {label}")
```

```python
print(df["Class"].value_counts())
sns.barplot(df["Class"].value_counts())
```

```python
sns.heatmap(df.isnull(), yticklabels = False, cbar = False, cmap = "coolwarm_r")
```

```python
X = df.drop(["Class"], axis = 1)
y = df["Class"]
```

```python
# Prerocessing, Normalizing the dataset
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

X_scaled_to_pd_dataframe = X_scaled
X_scaled_to_pd_dataframe = pd.DataFrame(X_scaled_to_pd_dataframe, columns = df.drop("Class", axis = 1).columns)
X_scaled_to_pd_dataframe.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.30, random_state = 12)
```

# Logistic Regression model

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
y_train
```

```python
y_pred
```

# Evaluating the model

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy: .5f}")
```

```python
print(classification_report(y_test, y_pred, target_names = ["Çerçevelik", "Ürgüp Sivrisi"]))
```
