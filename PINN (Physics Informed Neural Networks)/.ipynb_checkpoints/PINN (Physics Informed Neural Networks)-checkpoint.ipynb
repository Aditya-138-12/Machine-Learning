# Creating the Neural Network

```python
model = tf.keras.Sequential([
    tf.keras.layers.InputLayer(input_shape = (1,)),
    tf.keras.layers.Dense(32, activation = "tanh"),
    tf.keras.layers.Dense(32, activation = "tanh"),
    tf.keras.layers.Dense(2)
])
```

```python
def pinn_loss(model, t_input):
    t_input = tf.convert_to_tensor(t_input)
    t_input = tf.cast(t_input, dtype = tf.float32)
    with tf.GradientTape(persistent = True, watch_accessed_variables = True) as tape2:
        tape2.watch(t_input)
        with tf.GradientTape(persistent = True, watch_accessed_variables = True) as tape1:
            tape1.watch(t_input)
            pred = model(t_input)
            x = pred[:, 0:1]
            y = pred[:, 1:2]
        dx_dt = tape1.gradient(x, t_input)
        dy_dt = tape1.gradient(y, t_input)
    d2x_dt2 = tape2.gradient(dx_dt, t_input)
    d2y_dt2 = tape2.gradient(dy_dt, t_input)

    del tape1
    del tape2

    res_x = d2x_dt2
    res_y = d2y_dt2 + g

    physics_loss = tf.reduce_mean(tf.square(res_x)) + tf.reduce_mean(tf.square(res_y))
    return physics_loss
```

```python
def initial_condition_loss(model):
    t0 = tf.convert_to_tensor([[0.0]])
    pred0 = model(t0)
    x0_pred, y0_pred = pred0[0, 0], pred0[0, 1]

    with tf.GradientTape() as tape:
        tape.watch(t0)
        pred = model(t0)
        x = pred[:, 0:1]
        y = pred[:, 1:2]
    dx_dt = tape.gradient(x, t0)[0, 0]
    dy_dt = tape.gradient(y, t0)[0, 0]

    del tape

    loss_ic = (
        tf.square(x0_pred - x0) +
        tf.square(y0_pred - y0) +
        tf.square(dx_dt, vx0) +
        tf.square(dy_dt, vy0)
    )
    return loss_ic
```
