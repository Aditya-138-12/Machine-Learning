```python
from sklearn.model_selection import train_test_split
```

```python
new_X = data_encoded[["enginesize", "curbweight", "horsepower", "carlength", "fuelsystem_mpfi", "cylindernumber_four", "highwaympg", "drivewheel_fwd"]]
new_y = data_encoded["price"]

X = data_numeric[["enginesize", "curbweight", "horsepower", "carwidth", "carlength", "wheelbase"]]
y = data_numeric["price"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 12)

new_X_train, new_X_test, new_y_train, new_y_test = train_test_split(new_X, new_y, test_size = 0.2859, random_state = 12)
```

```python
from sklearn.linear_model import LinearRegression
#from sklearn.feature_selection import RFE
```

```python
lr = LinearRegression()
lr2 = LinearRegression()
```

```python
lr.fit(X_train, y_train)
lr2.fit(new_X_train, new_y_train)
# slecter = RFE(lr2, n_features_to_select = 9)
# slecter.fit(new_X_train, new_y_train)

# slected_features = new_X_train.columns[slecter.support_]
# print("Selected Features are: ", slected_features)
# lr2.fit(new_X_train[slected_features], new_y_train)
```

```python
print(lr.coef_, lr.intercept_)
```

```python
print(lr2.coef_, lr2.intercept_)
```

```python
coefficient_dataframe = pd.DataFrame(lr.coef_, X.columns, columns = ["Coef"])
coefficient_dataframe
```

```python
coefficient_dataframe = pd.DataFrame(lr2.coef_, new_X.columns, columns = ["Coef"])
coefficient_dataframe
```

```python
print("The R2 value on entire dataset is: ", lr.score(X, y)*100 , "%")
print("The R2 value on test dataset is: ", lr.score(X_test, y_test)*100 , "%")

print("The R2 value on entire dataset is: ", lr2.score(new_X, new_y)*100 , "%")
print("The R2 value on test dataset is: ", lr2.score(new_X_test, new_y_test)*100 , "%")
```

```python
predicted_values = lr.predict(X_test)
N_predicted_value = lr2.predict(new_X_test)
print(new_y_train,
N_predicted_value)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import math
from scipy.stats import norm
import numpy as np
```

```python
MAE = mean_absolute_error(new_y_test, N_predicted_value)
MSE = mean_squared_error(new_y_test, N_predicted_value)
RMSE = math.sqrt(MSE)
print("Mean Absolute Error is: ", MAE)
print("Mean Squared Error is: ", MSE)
print("Root Mean Squared Error is: ", RMSE)
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(new_y_test, N_predicted_value, alpha = 0.5, color = "blue")
plt.plot([new_y_test.min(), new_y_test.max()],[new_y_test.min(), new_y_test.max()], color = "red", linestyle = "-", alpha = 0.5)
#plt.plot([N_predicted_value.min(), N_predicted_value.max()], [N_predicted_value.min(), N_predicted_value.max()], color = "black", linestyle = "-", alpha = 0.5)
for actual, predicted in zip(new_y_test, N_predicted_value):
    plt.plot([actual, actual], [actual, predicted], color = "gray", linestyle = ":", alpha = 0.6)
    #plt.scatter([actual], [predicted], color = "yellow", alpha = 0.5)
    plt.scatter([actual], [actual], color = "green", alpha = 0.5)
plt.grid(True)
```

```python
plt.figure(figsize = (12, 10))
residuals = N_predicted_value - new_y_test
length_of_residuals = range(len(residuals))
plt.bar(length_of_residuals, residuals, color = "skyblue", alpha = 0.7)
plt.axhline(0, color = 'black', linestyle = '--', linewidth = 0.5, label = "Zerro Error line", alpha = 0.5)
```

```python
plt.figure(figsize=(12, 10))
sns.histplot(residuals, bins = 30, kde = True, stat = "density", color = "skyblue", label = "Normal Distribution Curve")
```

```python
import statsmodels.api as sm

X_sm = sm.add_constant(np.asarray(X_train).astype(float))
y_sm = np.asarray(y_train).astype(float)
  
model = sm.OLS(y_sm, X_sm).fit()

print(model.summary())

conf_intervals = model.conf_int()
print(conf_intervals)
```

## Shapely Values

```python
import shap
```

```python
print(new_X_test.dtypes)
print(new_X_test.head())
print(new_X_train.dtypes)

bool_cols = ["fuelsystem_mpfi", "cylindernumber_four", "drivewheel_fwd"]
new_X_train[bool_cols] = new_X_train[bool_cols].astype(int)
new_X_test[bool_cols] = new_X_test[bool_cols].astype(int)

explainer = shap.Explainer(lr2, new_X_train)
plt.figure(figsize = (12, 6))
shap_values = explainer(new_X_test)
shap.plots.beeswarm(shap_values)
shap.plots.bar(shap_values)
shap.plots.waterfall(shap_values[0])
```
