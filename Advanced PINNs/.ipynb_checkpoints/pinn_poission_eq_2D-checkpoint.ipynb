```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models 
import time

x = np.round(np.linspace(0, 1, 1024), 3)
y = np.round(np.linspace(0, 1, 1024), 3)
X1, Y1 = np.meshgrid(x, y)

T = np.sin(np.pi * X1)*np.sin(np.pi * Y1)

plt.figure()
contour = plt.contourf(X1, Y1, T, levels = 20)
plt.colorbar(contour)
plt.title('Poisson Equation Analytical Solution')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
plt.savefig('Poission_equation_anaytical_solution-l20.png', dpi=300, bbox_inches='tight')
```

```python
size = 256
x = np.round(np.linspace(0 , 1, size, dtype=np.float32), 3)
y = np.round(np.linspace(0, 1, size, dtype=np.float32), 3)
```

```python
# Neural Network architecture

def sinh_activation(x):
    return tf.math.sinh(x)

model = tf.keras.Sequential([
    layers.Input(shape=(2,)),
    layers.Lambda(lambda x: x * 2 - 1),
    layers.Dense(32, activation=sinh_activation),
    layers.Dense(64, activation='tanh'),
    layers.Dense(32, activation=sinh_activation),
    layers.Dense(1, activation=None)
])

model.summary()
optimizer = tf.keras.optimizers.Adam(learning_rate=1e-4, clipnorm=2.0)
```

```python
def differential_eq_loss_func(model, x, y):
    inputs = tf.convert_to_tensor(np.column_stack([x, y]))
    with tf.GradientTape(persistent=True) as tape2:
        tape2.watch(inputs)
        with tf.GradientTape(persistent=True) as tape1:
            tape1.watch(inputs)
            predictions = model(inputs)
        dt_dx = tape1.gradient(predictions, inputs)[:,0]
        dt_dy = tape1.gradient(predictions, inputs)[:,1]
        
    d2t_dx2 = tape2.gradient(dt_dx, inputs)[:,0]
    d2t_dy2 = tape2.gradient(dt_dy, inputs)[:,1]

    return tf.reduce_mean(((d2t_dx2 + d2t_dy2) + ((2 *( np.pi ** 2)) * tf.sin(np.pi * inputs[:,0]) * tf.sin(np.pi * inputs[:,1])))**2)

def boundary_eq_loss_func(model):
    # top case
    yT = np.full(size, 1, dtype=np.float32)
    xT = np.round(np.linspace(0, 1, size), 3)
    inputs = tf.convert_to_tensor(np.column_stack([xT, yT]))
    predictions = model(inputs)
    loss_top = tf.reduce_mean(predictions ** 2)

    # bottom case
    yB = np.full(size, 0, dtype=np.float32)
    xB = np.round(np.linspace(0, 1, size), 3)
    inputs = tf.convert_to_tensor(np.column_stack([xB, yB]))
    predictions = model(inputs)
    loss_btm = tf.reduce_mean(predictions ** 2)

    # Left Case
    xL = np.full(size, 0, dtype=np.float32)
    yL = np.round(np.linspace(0, 1, size), 3)
    inputs = tf.convert_to_tensor(np.column_stack([xL, yL]))
    predictions = model(inputs)
    loss_left = tf.reduce_mean(predictions ** 2)

    #right case
    xR = np.full(size, 1, dtype=np.float32)
    yR = np.round(np.linspace(0, 1, size), 3)
    inputs = tf.convert_to_tensor(np.column_stack([xR, yR]))
    predictions = model(inputs)
    loss_right = tf.reduce_mean(predictions ** 2)
    return (loss_top + loss_btm + loss_left + loss_right)/4.0
```

```python
from IPython.display import clear_output
epochs = 2500
lmb_bc = 14
lmb_pde = 20

pde_losses = []
bc_losses = []
total_losses = []

for epoch in range(epochs):
    start = time.time()

    epoch_loss = 0.0
    with tf.GradientTape() as tape:
        pde_loss = differential_eq_loss_func(model, x, y)
        bc_loss = boundary_eq_loss_func(model)
        total_loss = lmb_pde * pde_loss + bc_loss * lmb_bc
    gradient = tape.gradient(total_loss, model.trainable_variables)
    optimizer.apply_gradients(zip(gradient, model.trainable_variables))
    
    pde_losses.append(pde_loss.numpy())
    bc_losses.append(bc_loss.numpy())
    total_losses.append(total_loss.numpy())

    if epoch % 100 == 0:
        clear_output(wait=True)

        plt.figure(figsize=(8,5))
        plt.plot(total_losses, label="Total Loss")
        plt.plot(pde_losses, label="PDE Loss")
        plt.plot(bc_losses, label="BC Loss")

        plt.yscale("log")
        plt.xlim(0, epochs)
        plt.legend()
        plt.grid(True)
        plt.show()

    
    cost = time.time() - start
    print(f"Cost: {cost}, Epoch: {epoch}/{epochs}, pde_loss: {pde_loss.numpy():.6f}, bc_loss: {bc_loss.numpy():.6f}, total_loss: {total_loss.numpy():.6f}")
model.save('poission_equation.keras')
print(f"Model Saved as poission_equation.keras")
```

```python
from scipy.optimize import minimize

iteration = [0]
# Helper: flatten model weights to 1D numpy array
def get_weights_vector(model):
    weights = model.get_weights()
    return np.concatenate([w.flatten() for w in weights])

# Helper: set model weights from 1D numpy array
def set_weights_vector(model, weights_vector):
    weights = model.get_weights()
    new_weights = []
    idx = 0
    for w in weights:
        shape = w.shape
        size = w.size
        new_w = weights_vector[idx:idx+size].reshape(shape)
        new_weights.append(new_w)
        idx += size
    model.set_weights(new_weights)

# Loss function wrapper for L-BFGS
def loss_and_grad(weights_vector):
    set_weights_vector(model, weights_vector)

    with tf.GradientTape() as tape:
        # Calculate total loss here (your PDE + BC losses)
        pde_loss = differential_eq_loss_func(model, x, y)
        bc_loss = boundary_eq_loss_func(model)
        total_loss = lmb_pde * pde_loss + lmb_bc * bc_loss

    grads = tape.gradient(total_loss, model.trainable_variables)
    # Flatten gradients to 1D numpy array
    grads_vector = np.concatenate([g.numpy().flatten() for g in grads])
    if iteration[0] % 10 == 0:  # print every 10 steps
        print(f"L-BFGS iteration {iteration[0]}: total_loss = {total_loss.numpy():.6e}")
    iteration[0] += 1
    return total_loss.numpy().astype(np.float64), grads_vector.astype(np.float64)
    
# Run L-BFGS optimization
weights_init = get_weights_vector(model)

result = minimize(fun=loss_and_grad,
                  x0=weights_init,
                  jac=True,          # use gradient
                  method='L-BFGS-B',
                  options={'maxiter': 50000, 'maxcor': 50, 'maxfun': 50000, 'maxls': 50, 'ftol': 1.0 * np.finfo(float).eps})

# Set optimized weights back to model
set_weights_vector(model, result.x)
model.save("poission_equation_lbfgs.keras")
```
