# Solution of Steady-State Incompressible Navier Stokes Equation for Lid Driven Cavity case at Re = 100, U = 1, (x, y) in range [0, 1], 2-dimension

## Dimensionless Forms
## Mass Conservation 
### Ux + Vy = 0

##  Newton's Second Law
###   X-momentum  
####    U.Ux  +  V.Uy  = -Px + (1/Re).(Uxx + Uyy)
###   Y-momentum  
####    U.Vx  +  V.Vy  = -Py + (1/Re).(Vxx + Vyy)

## Boundary Conditions
### Top Wall
#### Top Lid (y=1) -> U = 1, V = 0
#### Bottom Wall (y=0) -> U = 0, V = 0
#### Left Wall (x=0) -> U = 0, V = 0
#### Right Wall (x=1) -> U = 0, V = 0

```python
# Data generation, raw points

import numpy as np
import tensorflow as tf

size = 512

x = np.round(np.linspace(0, 0.1, size, dtype=np.float32), 3,)
y = np.round(np.linspace(0, 0.1, size, dtype=np.float32), 3)
x = tf.convert_to_tensor(x)
y = tf.convert_to_tensor(y)
# x = tf.reshape(x, (-1, 1))
# y = tf.reshape(y, (-1, 1))
```

```python
# Top lid boundary conditions points
x_top = x
y_top = 0.1 * tf.ones(size, dtype=tf.float32)

# Bottom boundary conditions points
x_b = x
y_b = tf.zeros(size, dtype=tf.float32)

# Left boundry conditions points
x_left = tf.zeros(size, dtype=tf.float32)
y_left = y

# right boundary conditions points
x_right = 0.1 * tf.ones(size, dtype=tf.float32)
y_right = y

U_bc_top = tf.ones(size, dtype=tf.float32)
V_bc_top = tf.zeros(size, dtype=tf.float32)

U_bc_zero = tf.zeros(size, dtype=tf.float32)
V_bc_zero = tf.zeros(size, dtype=tf.float32)
```

### PINN architecture
#### Total 4 hidden layers, this is a Fully Connected Network with tanh Activation function in the hidden layers, the output layer does not have any functions

```python
from tensorflow.keras import layers, models
from tensorflow.keras.utils import plot_model
import time

model = tf.keras.Sequential([
    layers.Input(shape=(2, )),
    layers.Lambda(lambda x: x * 2.0 - 1.0),
    layers.Dense(32, activation="tanh"),
    # layers.Dense(64, activation="tanh"),
    layers.Dense(32, activation="tanh"),
    layers.Dense(3)
])

plot_model(
    model,
    to_file='ns_graph.png',
    show_shapes=True,
    show_layer_names=True,
    expand_nested=True,
    dpi=100
)
```

### Navier Stoke's Equation Loss functions

```python
def mass_conservation_loss(model, x, y):
    inputs = tf.convert_to_tensor(np.column_stack([x, y]))
    inputs = tf.Variable(inputs)
    with tf.GradientTape(persistent=True) as tape:
        # tape.watch(x)
        # tape.watch(y)
        predictions = model(inputs)
        U = predictions[:,0]
        V = predictions[:,1]
        P = predictions[:,2]
    U_x = tape.gradient(U, inputs)[:,0]
    V_y = tape.gradient(V, inputs)[:,1]

    return tf.reduce_mean((U_x + V_y)**2)

def x_momentum_loss(model, Re, x, y):
    inputs = tf.convert_to_tensor(np.column_stack([x, y]))
    #inputs = tf.Variable(inputs)
    with tf.GradientTape(persistent=True) as tape2:
        tape2.watch(inputs)
        with tf.GradientTape(persistent=True) as tape1:
            tape1.watch(inputs)
            predictions = model(inputs)
            U = predictions[:,0]
            V = predictions[:,1]
            P = predictions[:,2]
        U_x = tape1.gradient(U, inputs)[:,0]
        V_x = tape1.gradient(V, inputs)[:,0]
        P_x = tape1.gradient(P, inputs)[:,0]
        U_y = tape1.gradient(U, inputs)[:,1]
    
    U_y_y = tape2.gradient(U_y, inputs)[:,1]
    U_x_x = tape2.gradient(U_x, inputs)[:,0]

    eqn = -(U * U_x) - (V * U_y) - P_x + ((1/Re) * (U_x_x + U_y_y))
    return tf.reduce_mean(eqn**2)

def y_momentum_loss(model, Re, x, y):
    inputs = tf.convert_to_tensor(np.column_stack([x, y]))
    #inputs = tf.Variable(inputs)
    with tf.GradientTape(persistent=True) as tape2:
        tape2.watch(inputs)
        with tf.GradientTape(persistent=True) as tape1:
            tape1.watch(inputs)
            predictions = model(inputs)
            U = predictions[:,0]
            V = predictions[:,1]
            P = predictions[:2]
        V_x = tape1.gradient(V, inputs)[:,0]
        V_y = tape1.gradient(V, inputs)[:,1]
        P_y = tape1.gradient(P, inputs)[:,1]
    V_y_y = tape2.gradient(V_y, inputs)[:,1]
    V_x_x = tape2.gradient(V_x, inputs)[:,0]

    eqn = -(U * V_x) - (V * V_y) - P_y + ((1/Re) * (V_x_x + V_y_y))
    return tf.reduce_mean(eqn**2)

def boundary_loss(model):
    # top lid loss
    predictions = model(tf.convert_to_tensor(np.column_stack([x_top, y_top])))
    U_bc_pred = predictions[:,0]
    V_bc_pred = predictions[:,1]
    loss_top = tf.reduce_mean((U_bc_pred - U_bc_top)**2 + (V_bc_pred - V_bc_top)**2)

    # bottom wall loss
    predictions = model(tf.convert_to_tensor(np.column_stack([x_b, y_b])))
    U_bc_pred = predictions[:,0]
    V_bc_pred = predictions[:,1]
    loss_bottom = tf.reduce_mean((U_bc_pred - U_bc_zero)**2 + (V_bc_pred - V_bc_zero)**2)

    # Left wall loss
    predictions = model(tf.convert_to_tensor(np.column_stack([x_left, y_left])))
    U_bc_pred = predictions[:,0]
    V_bc_pred = predictions[:,1]
    loss_left = tf.reduce_mean((U_bc_pred - U_bc_zero)**2 + (V_bc_pred - V_bc_zero)**2)

    # right wall loss
    predictions = model(tf.convert_to_tensor(np.column_stack([x_right, y_right])))
    U_bc_pred = predictions[:,0]
    V_bc_pred = predictions[:,1]
    loss_right = tf.reduce_mean((U_bc_pred - U_bc_zero)**2 + (V_bc_pred - V_bc_zero)**2)

    return loss_top + loss_bottom  + loss_left + loss_right

def pressure_loss(model, x, y):
    inputs = tf.convert_to_tensor(np.column_stack([x, y]))
    predictions = model(inputs)
    P = predictions[:,2]
    return tf.reduce_mean((P - 0)**2)  # To make the network calculate gradient of P too!
    
def initial_loss(model):
    pass
```

```python
# Training loop

batch_size =  64
epochs = 10000
Re = 100.0

lmb_pde = 14
lmb_bc = 15
lmb_pr = 3

# dataset = tf.stack([x, y], axis=0)
dataset = tf.data.Dataset.from_tensor_slices((x, y))
dataset = dataset.shuffle(buffer_size=len(x))
dataset = dataset.batch(batch_size=batch_size)
dataset
```

```python
from IPython.display import clear_output
optimizer = tf.keras.optimizers.Adam(learning_rate=1e-3)

pde_losses = []
bc_losses = []
pr_losses = []
total_losses = []

for epoch in range(epochs):
    start = time.time()
    epoch_loss = 0.0

    for x, y in dataset:
        with tf.GradientTape() as tape:
            pde_loss = mass_conservation_loss(model, x, y) + x_momentum_loss(model, Re, x, y) + y_momentum_loss(model, Re, x, y)
            bc_loss = boundary_loss(model)
            pr_loss = pressure_loss(model, x, y)
            total_loss = (lmb_pde * pde_loss) + (lmb_bc * bc_loss) + (lmb_pr * pr_loss)

        gradients = tape.gradient(total_loss, model.trainable_variables)
        optimizer.apply_gradients(zip(gradients, model.trainable_variables))

    pde_losses.append(pde_loss.numpy())
    bc_losses.append(bc_loss.numpy())
    pr_losses.append(pr_loss.numpy())
    total_losses.append(total_loss.numpy())
        
    if epoch % 100 == 0:
        clear_output(wait=True)

        plt.figure(figsize=(15,8))
        plt.plot(total_losses, label="Total Loss", alpha=0.5)
        plt.plot(pde_losses, label="PDE Loss", alpha=0.5)
        plt.plot(bc_losses, label="BC Loss", alpha=0.5)
        plt.plot(pr_losses, label="PR Loss", alpha=0.5)

        plt.yscale("log")
        plt.xlim(0, epochs)
        plt.legend()
        plt.grid(True)
        plt.show()

        
    cost = time.time() - start
    print(f"\nTime taken: {cost}\n Epoch: {epoch + 1}/{epochs}, pde_loss: {pde_loss.numpy():.6f}, bc_loss: {bc_loss.numpy():.6f}, pr_loss: {pr_loss.numpy():.6f}, total_loss: {total_loss.numpy():.6f}\n")

print(f"model saved as navier_stokes_incompressible_steady_state_lid_drive_cavity_2d.keras")
#model.save('navier_stokes_incompressible_steady_state_lid_drive_cavity_2d.keras')
```

```python
from scipy.optimize import minimize

#model1 = tf.keras.models.load_model('navier_stokes_incompressible_steady_state_lid_drive_cavity_2d.keras', safe_mode=False)
iteration = [0]
# Helper: flatten model weights to 1D numpy array
def get_weights_vector(model):
    weights = model.get_weights()
    return np.concatenate([w.flatten() for w in weights])

# Helper: set model weights from 1D numpy array
def set_weights_vector(model, weights_vector):
    weights = model.get_weights()
    new_weights = []
    idx = 0
    for w in weights:
        shape = w.shape
        size = w.size
        new_w = weights_vector[idx:idx+size].reshape(shape)
        new_weights.append(new_w)
        idx += size
    model.set_weights(new_weights)

# Loss function wrapper for L-BFGS
def loss_and_grad(weights_vector):
    set_weights_vector(model, weights_vector)

    with tf.GradientTape() as tape:
        # calculation of total loss
        pde_loss = mass_conservation_loss(model, x, y) + x_momentum_loss(model, Re, x, y) + y_momentum_loss(model, Re, x, y)
        bc_loss = boundary_loss(model)
        pr_loss = pressure_loss(model, x, y)
        total_loss = lmb_pde * pde_loss + lmb_bc * bc_loss + lmb_pr * pr_loss

    grads = tape.gradient(total_loss, model.trainable_variables)
    # flatten gradients to 1D numpy array
    grads_vector = np.concatenate([g.numpy().flatten() for g in grads])
    if iteration[0] % 10 == 0:  # print every 10 steps
        print(f"L-BFGS iteration {iteration[0]}: total_loss = {total_loss.numpy():.6e}")
    iteration[0] += 1
    return total_loss.numpy().astype(np.float64), grads_vector.astype(np.float64)
    
# Run L-BFGS optimization
weights_init = get_weights_vector(model)

result = minimize(fun=loss_and_grad,
                  x0=weights_init,
                  jac=True,          # use gradient
                  method='L-BFGS-B',
                  options={'maxiter': 50000, 'maxcor': 50, 'maxfun': 50000, 'maxls': 50, 'ftol': 1.0 * np.finfo(float).eps})

# Set optimized weights back to model
set_weights_vector(model, result.x)
#model.save("navier_stokes_incompressible_steady_state_lid_drive_cavity_lbfgs.keras")
```
